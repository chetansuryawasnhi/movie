```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies=movies.merge(credits,on="title")
```

```python
# genres , id , keywords , title , overview  , cast , crew

```

```python
main=movies[['genres','id','keywords', 'title','overview','cast','crew']]
```

```python
main
```

```python
main.head()
```

```python
main.isnull().sum()
```

```python
main.dropna(inplace=True)
```

```python
main.isnull().sum()
```

```python
main.duplicated().sum()
```

```python
main.genres
```

```python
new=[]
for i in main.genres:
    # print(eval(i))
    a=[]
    for j in range(len(eval(i))):
        a.append(eval(i)[j]['name'])
    new.append(a)
print(new)
main.genres=new
```

```python
main.genres
```

```python
def convert(obj):
    l=[]
    for i in eval(obj):
        l.append(i['name'])
    return l
```

```python
main['keywords']=main['keywords'].apply(convert)
main['cast']=main['cast'].apply(convert)
# main['crew']=main['crew'].apply(convert)
```

```python
for i in range(len(main["cast"])):
    main["cast"][i]=main["cast"][i][:3]
```

```python


def fetch(obj):
    l=[]
    
    for i in eval(obj):
        if i['job']=="Director":
            l.append(i['name'])
    return l
```

```python
main['crew']=main['crew'].apply(fetch)
```

```python
main['crew']
```

```python
main.overview=main.overview.apply(lambda x:x.split())
```

```python
main.head()
```

```python
main.genres=main.genres.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
main.crew=main.crew.apply(lambda x:[i.replace(" ","") for i in x])
main.cast=main.cast.apply(lambda x:[i.replace(" ","") for i in x])
main.keywords=main.keywords.apply(lambda x:[i.replace(" ","") for i in x])

```

```python
main.head()
```

```python
main['tag']=main.genres+main.keywords+main.overview+main.cast+main.crew
```

```python
main['tag'][0]
```

```python
movies=main[['id','title','tag']]
```

```python
movies.tag=movies.tag.apply(lambda x:" ".join(x))
```

```python
movies['tag']
```

```python
movies.tag=movies.tag.apply(lambda x:x.lower())
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies.tag).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies.tag=movies.tag.apply(stem)
```

```python
movies
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vectors)
```

```python
sim.shape
```

```python
sim[1]
```

```python
def reco(node):
    ind=movies[movies['title']==node].index[0]

    dis=list(enumerate(sim[ind]))
    dis.sort(key=lambda x:x[1],reverse=True)
    top_5= dis[1:6]
    for i in dis[1:6]:
        # print(i[0])
        print(movies.iloc[i[0]].title)
    
reco('Batman')
```
